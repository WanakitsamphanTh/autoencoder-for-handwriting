```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
from scipy.stats import skew,kurtosis
```

```python
chars = ['a','nu','ti','ke','yo','yorozu']
jpchars = dict(zip(chars,['あ','ぬ','チ','ケ','世','万']))
```

```python
xl = {}
for c in chars:
    f = pd.read_excel('ds1-loss.xlsx', sheet_name=c, engine='openpyxl', index_col = 0)
    f.rename(columns={0:'File',1:'Loss'},inplace=True)
    xl[c] = f
```

```python
ds1_losses = {}
for c in chars:
    ds1_losses[c] = xl[c]['Loss']
```

```python
xl2 = {}
for c in chars:
    f = pd.read_excel('ds2-loss.xlsx', sheet_name=c, engine='openpyxl', index_col = 0)
    f.rename(columns={0:'File',1:'Loss'},inplace=True)
    xl2[c] = f
```

```python
ds2_losses = {}
for c in chars:
    ds2_losses[c] = xl2[c]['Loss']
```

```python
for c in chars:
    print("%s : DS1_max = %f, DS2_max = %f" % (c, ds1_losses[c].max(), ds2_losses[c].max()))
```

```python
c = 'a'
plt.hist(ds1_losses[c],weights=np.ones(len(ds1_losses[c])) / len(ds1_losses[c]), range = [0.0,.6], color='lightgreen', ec='black', bins=20, label = 'By Natives (DS1)\nmean = %.3f\n' % (np.mean(ds1_losses[c])) , alpha=0.5)
plt.hist(ds2_losses[c],weights=np.ones(len(ds2_losses[c])) / len(ds2_losses[c]),range = [0.0,.6], color='red', ec='black', bins=20, label = 'By Foreigner Novices (DS2)\nmean = %.3f' % (np.mean(ds2_losses[c])),alpha=0.5)
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.ylim(0.0,0.8)
plt.title(jpchars[c], fontname = 'Yu Gothic')
plt.legend()

plt.savefig(c + '-hist.png')
```
